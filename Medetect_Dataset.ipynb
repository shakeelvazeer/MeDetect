```python
import matplotlib.pyplot as plt
import os
```

```python
train_directory = 'F:/Class/FYP/Data/maindataset/Train'
validation_directory = 'F:/Class/FYP/Data/maindataset/Val'

class_labels = sorted(os.listdir(train_directory))

print(class_labels)
```

```python
# Count the number of images in each class
pill_names = []
num_images_per_pill = []
for pill_name in os.listdir(train_directory):
    class_path = os.path.join(train_directory, pill_name)
    if os.path.isdir(class_path):
        pill_names.append(pill_name)
        num_images_per_pill.append(len(os.listdir(class_path)))

# Plot the distribution
plt.figure(figsize=(10, 6))
plt.bar(pill_names, num_images_per_pill)
plt.xlabel('Pill Name')
plt.ylabel('Number of images')
plt.xticks(rotation=90)
plt.title('Distribution of Images Across Classes')
plt.show()
```
